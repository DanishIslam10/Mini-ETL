# Remove Rows having Fare = 0

Why do some passengers have Fare = 0?

Possible reasons:

They were crew members or working staff

They were family members traveling under the same shared ticket

Children or dependents traveling free

Company-sponsored travelers

```python
filt = df['Fare'] == 0
df[filt].shape[0] #there are 15 people who are travelling free
```

```python
filt = df['Fare'] > 0
df = df[filt] #only kept rows having Fare greater than zero
```

# Fill Missing Ages with median age group wise

We made group with 'Pclass' and 'Sex' to fill the ages realistically.
Age depends highly on 'Pclass' and 'Sex'.
Females which are 1st class passengers are generally older.
Males which are 3rd class passengers are generally younger.
So we need to define the groups with 'Pclass' and 'Sex' to have more realistic and precise age.

```python
filled_ages = df.groupby(['Pclass','Sex'])['Age'].transform(lambda x: x.fillna(x.median()))
```

```python
df['Age'] = filled_ages
```

```python
df['Age'].isna().sum()
```

# Fill Embarked Column with the most common port

```python
df["Embarked"] = df.groupby('Pclass')['Embarked'].transform(lambda x: x.fillna(x.mode()[0]))
```

```python
df['Embarked'].isna().sum()
```

# Fill Missing Cabins

In the data set, we are given Cabin numbers (cabin deck + actual room number) , but it is not relevant
because each cabin number is unique and with these unique values no pattern can be observed.
Instead, we just need to extract the cabin deck (floor) so we can observe the pattern of survival.
Higher the deck more chances to take the lifeboat and survive, and higher deck are expensive are these 
people have more priority to be saved.
So instead of finding the missing cabins which is not possible, we should extract the cabin dekc (first letter of cabin)
and make a new column named 'Cabin Deck'.

```python
df['Cabin Deck'] = df['Cabin'].str[0]
```

```python
df.groupby('Pclass')['Cabin Deck'].agg(lambda x: x.mode()[0])
```

```python
df.head(50)
```

## Now fill missing value in Cabin Deck with 'Unknown'

üö¢ Why We Do Not Fill Missing Cabin Deck Values

The Cabin feature has a very high number of missing values ‚Äî most passengers do not have a recorded cabin number.
Instead of treating missing values as random or filling them with a substitute (like the mode), the absence of a cabin entry itself carries meaningful information.

üîç Observations

1st class passengers are more likely to have a cabin recorded
‚Üí They had better accommodation and were more likely to survive.

2nd and especially 3rd class passengers mostly have missing cabin information
‚Üí These passengers had lower survival probability and poorer access to lifeboats.

So, Missing Cabin = lower socio-economic status, which correlates with survival.

```python
df['Cabin Deck'] = df['Cabin Deck'].fillna("Unknown")
```

```python
filt = (df['Pclass'] == 3) & (df['Cabin Deck'] == "Unknown")
df[filt].shape[0]
```
